```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("movies.csv")
```

```python
print("Original Dataset")
print(df)
```

```python
df.drop_duplicates()
df.dropna()
df.drop(columns=['Poster_Link'],inplace=True)
df.head()
```

```python
if 'Genre' in df.columns:
    top_genres=df['Genre'].value_counts().head()
    plt.figure(figsize=(10,5))
    top_genres.plot(kind='bar', color='red')
    plt.title("Top 10 Most Common Genres")
    plt.xlabel("Genre")
    plt.ylabel("Count")    plt.grid(True)
    plt.tight_layout()
    plt.show()
else:
    print("No 'genre' column found for visualization")
```

```python
if 'IMDB_Rating' in df.columns:
    df['IMDB_Rating'].plot(kind='hist',bins=10, color='green', edgecolor='black')
    plt.title("Distribution of movie ratings")
    plt.xlabel("Rating")
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()
```

```python
if 'Released_Year' in df.columns:
    top_years=df['Released_Year'].value_counts().head(10)
    plt.figure(figsize=(8,5))
    top_years.plot(kind='bar', color='violet')
    plt.title("Top 10 Most Common Release Years")
    plt.xlabel(" Release Year")
    plt.ylabel("Number of Movies")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
